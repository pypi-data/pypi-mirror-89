# Linear Regression Example

```python
from pyLearnAlgorithms import 
from pyLearnAlgorithms.graph_view import GraphView
from pyLearnAlgorithms.linear_regression import LinearRegression
```

```python
dataset = LoadData('../tests/datasets', 'WaterLevel.mat')
```

```python
data = dataset.load_mat()
```

```python
X, y = data['X'], data['y'][:, 0]
Xtest, ytest = data['Xtest'], data['ytest'][:, 0]
Xval, yval = data['Xval'], data['yval'][:, 0]
```

```python
dataset_graph = GraphView(X, y)
dataset_graph.linear_regression_data('Mudança no nível da água (x)',
                                     'Água fluindo para fora da barragem (y)', 
                                     'Análise do nível de água em uma barragem')
```

```python
regressor = LinearRegression(X, y)
```

```python
J, grad = regressor.linearRegCostFunction()
```

```python
J
```

```python
grad
```

```python
optimize_values = regressor.trainLinearReg()
```

```python
optimize_values.fun
```

```python
optimize_values.x
```

```python
dataset_graph.model_linear('Mudança no nível da água (x)',
                           'Água fluindo para fora da barragem (y)', 
                           'Análise do nível de água em uma barragem')
```

```python
train_error, val_error = regressor.learningCurve(Xval, yval)
```

```python
print('# Training Examples\tTraining Error\t\tCross Validation Error')
for i in range(y.size):
    print('{}\t\t\t{}\t{}'.format(i+1, train_error[i], val_error[i]))
```

```python
dataset_graph.learning_curve_linear(Xval, yval)
```

```python
pred = regressor.predict(Xtest)
```

```python
dataset_graph.predicted_values(Xtest, ytest, pred)
```

```python
J, grad = regressor.normal_equation()
```

```python
dataset_graph.model_linear_normal_equation('Mudança no nível da água (x)',
                                           'Água fluindo para fora da barragem (y)', 
                                           'Análise do nível de água em uma barragem')
```

```python
pred_NE = regressor.predict_normal_equation(Xtest)
```

```python
dataset_graph.predicted_values(Xtest, ytest, pred_NE)
```
